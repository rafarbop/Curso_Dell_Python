**Visualização Básica em Barras de dados**

Usando a biblioteca matplotlib, usamos o método `bar` do módulo pyplot.  
Passamos para esse método um sequência de categorias e os valores associados

```python
pessoas = ['Pedro', 'Rafael','Roberto','Maria']
ano_nascimento = np.array([1987,1988,1999,1992])
idades = 2021-ano_nascimento
plot.figure(figsize=(8,5))
plot.bar(pessoas,idades,color='#008585')
plot.xlabel('Pessoas')
plot.ylabel('Idades')
plot.title('Testando Gráfico de Barras com Matplotlib')
plot.show()
```

```python
df = pd.read_csv('Empenhado_UO158953_2021.csv',sep=';',encoding="UTF-8")
if df['Valor Empenhado'].dtype != 'float':
    df['Valor Empenhado'] = df['Valor Empenhado'].map(lambda x: x.replace('.','').replace(',','.'))
    df['Valor Empenhado'] = df['Valor Empenhado'].astype('float')
df.sort_values(by='Mês Ano',inplace=True)
df.index = range(0,len(df.index))
```

```python
df
```

```python
plot.figure(figsize=(12,8))
plot.bar(df['Mês Ano'],df['Valor Empenhado'],color='#256585')
plot.ylabel('Valores Empenhados')
plot.xlabel('Mês/Ano')
plot.title('Valores Empenhados em 2021 - UG 158953 - IFCE Campus Tabuleiro do Norte')
plot.show()
```
