## Aula 3.3 - Visualizando séries temporais

```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn
```

```python
room = pd.read_csv('Dados_Oficina_03.3.txt')
room.head()
```

```python
plot.figure(figsize=(10,5))
plot.title('Variação da temperatura ao longo do tempo')
seaborn.scatterplot(x="date", y="Temperature", data=room)
plot.show()
```

```python
room['date'] = pd.to_datetime(room['date'])
plot.figure(figsize=(10,5))
plot.title('Variação da temperatura ao longo do tempo')
seaborn.lineplot(x="date", y="Temperature", data=room)
plot.show()
```
