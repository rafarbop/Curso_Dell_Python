**Gráficos de Pizza - Pie Charts**

A representação de dados proporcionais pode ser realizada, de forma simples, através de gráficos circulares, onde cada *Fatia da Pizza* representa a proporção de cada item.
Pode se feito com a bilioteca matplotlib de forma simples com a função **pie**

```python
animais = ['Onça','Jacaré','Girafa','Oragotango','Rinoceronte','zebra']
quantidade = [3,2,2,4,1,5]
plot.figure(figsize=(8,8))
plot.title('Animais em um Zoológico!')
plot.pie(x=quantidade,labels=animais,autopct='%1.2f%%')
plot.gcf().set_facecolor('#f2f2f2')
plot.style.use('seaborn-dark')
plot.show()
```

```python
plot.title('Animais em um Zoológico!')
plot.pie(x=quantidade,labels=animais,autopct='%1.2f%%',pctdistance=0.85,startangle=90)
p = plot.gcf()
p.gca().add_artist(plot.Circle((0,0),0.7,color='white'))
plot.show()
```
