Esse conjunto de dados chama-se “mpg” e é bastante famoso em estatística, sendo disponibilizado, diretamente, pela biblioteca seaborn sem a necessidade de baixar os dados para a sua máquina. 1. Importe as bibliotecas pandas, matplotlib e seaborn, usando a palavra mágica inline, para exibir as visualizações no notebook. Carregue os dados, que podem ser carregados pelo seaborn, utilizando a função load_dataset(‘mpg’), em um DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn
setmpg = seaborn.load_dataset("mpg")
setmpg.head()
```

2. Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’). Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem? Você pode escolher a técnica que julgar mais adequada!

```python
plot.figure(figsize=(14,7))
plot.title("Relação entre Aceleração e o Consumo")
seaborn.scatterplot(x='mpg',y='acceleration',hue='origin',size='origin',sizes=(40,160),data=setmpg)
plot.show()
seaborn.lmplot(x='mpg',y='acceleration',col='origin',data=setmpg)
plot.show()
```

Resposta: De acordo com os gráficos acima, verifica-se que a intepretação pode ser de uma correlação positiva muito fraca ou até mesmo não ter correlação entre a acelaração e o consumo.
A correlação é melhor verificada no Japão, pois nos outros países verifica-se a situação de correlação muito fraca ou inexistente, a depender da interpretação.

3. Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year). Existe alguma relação entre as três variáveis?

```python
plot.figure(figsize=(14,7))
plot.title("Relação entre Aceleração e o Consumo")
seaborn.scatterplot(x='mpg',y='acceleration',hue='model_year',size='model_year',sizes=(40,160),data=setmpg)
plot.show()
```

Resposta: De acordo com o gráfico acima, verifica-se que de acordo os carros mais novos (de modelos mais recentes), possuem melhor consumo de combustível do que modelos mais antigos. Referente a acelaração, não é possível verificar uma correlação com o ano.
