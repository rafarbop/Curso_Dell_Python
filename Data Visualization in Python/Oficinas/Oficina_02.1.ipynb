```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn
```

1. Importe as bibliotecas pandas, matplotlib, seaborn e use o comando mágico %matplotlib inline para exibir as visualizações no notebook. Carregue os dados em um DataFrame. Os dados estão disponibilizados no link:

```python
movies_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt'
df = pd.read_json(movies_uri)
df.head()
```

2. Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?

```python
plot.figure(figsize=(12,8))
seaborn.set_theme(style='darkgrid')
seaborn.color_palette("rocket")
seaborn.barplot(x='Budget_M',y='Film',data=df,palette='dark',order=df.sort_values('Budget_M').Film)
plot.show()
```

```python
max_budget = df['Film'][df['Budget_M'].idxmax()]
min_budget = df['Film'][df['Budget_M'].idxmin()]

print('Verificando o gráfico confirmamos facilmente que:')
print(f'\tO filme com maior orçamento é:\t{max_budget}')
print(f'\tO filme com menor orçamento é:\t{min_budget}')
```

3. Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?

```python
plot.figure(figsize=(12,8))
seaborn.set_theme(style='darkgrid')
seaborn.color_palette("rocket")
seaborn.barplot(x='Worldwide_Gross_M',y='Film',data=df,palette='dark',order=df.sort_values('Worldwide_Gross_M').Film)
plot.show()
```

```python
max_gross = df['Film'][df['Worldwide_Gross_M'].idxmax()]
min_gross = df['Film'][df['Worldwide_Gross_M'].idxmin()]

print('Verificando o gráfico confirmamos que:')
print(f'\tO filme com maior Bilheteria é:\t{max_gross}')
print(f'\tO filme com menor Bilheteria é:\t{min_gross}')
```

4. Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano. Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?

```python
plot.figure(figsize=(14,10))
seaborn.set_theme(style='darkgrid')
seaborn.color_palette("rocket")
seaborn.countplot(x='Year',hue='Genre',data=df,palette='dark')
plot.show()
```

Pelo gráfico acima verificamos que nos anos de **2008** e **2010** forma feitos menos filmes de Ação (2 filmes em cada ano)

E foram feitos menos filmes de animação no ano de **2011**.
