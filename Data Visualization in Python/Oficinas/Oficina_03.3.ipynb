```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn
shampoo = pd.read_csv('dadosoficina07.txt')
shampoo
```

```python
plot.figure(figsize=(14,8))
plot.title('Venda de Shampoo')
seaborn.lineplot(x="Month", y="Sales", data=shampoo)
plot.show()
```

Resposta: Pela leitura do gráfico, percebe-se uma tendência de crescimento nas vendas, sendo maior no último ano e menor no primeiro ano.
