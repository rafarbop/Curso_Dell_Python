## Aula 4.1 - Visualizando distribuições entre categorias

```python
import pandas as pd
import matplotlib.pyplot as plot
import seaborn
```

```python
titanic = seaborn.load_dataset("titanic")
titanic
```

```python
plot.figure(figsize=(12,6))
plot.title('Idade de passageiros em cada classe - Titanic')
seaborn.stripplot(x='class',y='age',data=titanic)
plot.show()
```

```python
plot.figure(figsize=(12,6))
plot.title('Idade de passageiros em cada classe - Titanic')
seaborn.swarmplot(x='class',y='age',data=titanic)
plot.show()
```

```python
seaborn.catplot(x="class", y="age", hue = 'sex', col = 'survived', kind = 'swarm', data=titanic)
```

**Gráficos de Caixa**

```python
plot.figure(figsize=(12,6))
plot.title('Idade de passageiros em cada classe - Titanic')
seaborn.boxplot(x='class',y='age',data=titanic)
plot.show()
```

```python
plot.figure(figsize=(12,6))
plot.title('Idade de passageiros em cada classe - Titanic')
seaborn.boxplot(x='class',y='age',hue='sex',data=titanic)
plot.show()
```

```python
plot.figure(figsize=(12,6))
plot.title('Idade de passageiros em cada classe - Titanic')
seaborn.boxenplot(x='class',y='age',hue='sex',data=titanic)
plot.show()
```

```python
seaborn.catplot(x="class", y="age", hue = 'sex', col = 'survived', kind = 'boxen', data=titanic)
```

**Gráfico de Violino**

```python
plot.figure(figsize=(12,6))
plot.title('Idade de passageiros em cada classe - Titanic')
seaborn.violinplot(x='class',y='age',data=titanic)
plot.show()
```

```python
plot.figure(figsize=(12,6))
plot.title('Idade de passageiros em cada classe - Titanic')
seaborn.swarmplot(x='class',y='age',color='black',size=5,alpha=0.2,data=titanic)
seaborn.violinplot(x='class',y='age',data=titanic)
plot.show()
```

```python
plot.figure(figsize=(12,6))
plot.title('Idade de passageiros em cada classe - Titanic')
seaborn.violinplot(x='class',y='age',hue='sex',split=True,data=titanic)
plot.show()
```

```python
seaborn.catplot(x='class',y='age',hue='sex',col='survived',kind='violin',split=True,data=titanic)
```
