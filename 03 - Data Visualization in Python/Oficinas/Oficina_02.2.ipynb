```python
import pandas as pd
import matplotlib.pyplot as plot
```

1. Importe as bibliotecas pandas e matplotlib, usando a palavra mágica inline para exibir as visualizações no notebook; e carregue os dados, que estão disponibilizados neste link: [dados](https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt), em um DataFrame. Assim, você pode carregar os dados pela URI ou salvar em sua máquina e carregá-los localmente;

```python
movies_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt'
df = pd.read_csv(movies_uri)
df.head()
```

2. Crie uma visualização que mostre a proporção de vendas global entre gêneros de jogos da empresa Nintendo;

```python
games_nintendo = df[df['Publisher'] == 'Nintendo']
games_nintendo = games_nintendo.groupby('Genre').sum()
games_nintendo.sort_values('Global_Sales',ascending=False)
proporcion_nintendo = games_nintendo['Global_Sales']*100/games_nintendo['Global_Sales'].sum()
plot.figure(figsize=(15,8),)
plot.title('Games by Genre - Nintendo')
plot.bar(x=proporcion_nintendo.index,height=proporcion_nintendo)
plot.grid(axis='y',linestyle='dotted',linewidth=1.4)
plot.show()
```

3. Crie uma visualização que mostre a proporção de quantidade de jogos por plataforma a partir do ano de 2015.

```python
games_after_2015 = df[df['Year'] >= 2015]
games_after_2015 = games_after_2015.groupby('Platform').count()
games_after_2015.sort_values('Name',ascending=False)
proporcion_after_2015 = games_after_2015['Name']*100/games_after_2015['Name'].sum()
plot.figure(figsize=(15,8),)
plot.title('Games by Platform - After 2015')
plot.bar(x=proporcion_after_2015.index,height=proporcion_after_2015)
plot.grid(axis='y',linestyle='dotted',linewidth=1.4)
plot.show()
```
